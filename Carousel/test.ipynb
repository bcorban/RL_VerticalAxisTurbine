```python
import torch
import numpy as np
import time
import threading
import RL_ 
import gym
import gclib
```

```python
g = gclib.py()
c = g.GCommand
g.GOpen("192.168.255.200 --direct -s ALL")
env=gym.make("RL_/CustomEnv-v0")

env.reset()
```

```python
def continuously_read():
    t=time.time()
    while time.time()-t<5:
        env.read_state()
```

```python
daemon=threading.Thread(target=continuously_read,daemon=True,name="state_reader")
```
